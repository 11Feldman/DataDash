## **1. ¿Qué son los histogramas?**

Un histograma es la representación gráfica en forma de barras, que simboliza la distribución de un conjunto de datos. Muestra la forma de los valores, o la distribución, de una variable continua. En otras palabras, un histograma es un gráfico o diagrama, que muestra el número de veces que se repiten cada uno de los resultados (o categorías), cuando se realizan mediciones sucesivas.

## **ENTONCES...** **¿Cómo entra el histograma acá?**

Primero debería ordenar los datos de menor a mayor para poder agruparlos. Agrupar los datos es importante porque si no se agrupan es muy dificil llegar a conclusiones evaluando los 3000 resultados distntos. Agrupar ayuda a simplificar el análisis y a la vez a poder trasladar las conclusiones a otras muestras. 

Cada grupo determinado por un rango de valores es representado por una barra. Los rangos de las barras se conocen como clases o intervalos. La altura de las barras muestra la frecuencia de los valores de datos en cada una. Por ejemplo, al crear el histograma de mediana de ingreso, podríamos decidir clasificarlos en XX cantidad de barras. La altura de las barras nos mostraría la cantidad de personas en cada uno de las barras = cada uno de los grupos.


```python

fig, ax = plt.subplots(figsize=(12, 4)) #figsize=(width, height) por pulgadas

histograma = sns.histplot(data= tabla, x='total_bedrooms', bins= 20)
ax.set_xlabel('Cantidad de cuartos')
ax.set_ylabel('Frecuencia')
ax.set_title('Distribucion de cantidad de cuartos en California');
```

## 3.**¿Qué tipo de dato corresponde a un histograma?**
a. Diferencia en los dos ejes

En un histograma el eje de las x (o abscisas) consiste del rango en el cual se encuentran los datos. Ahora, las bases de los rectángulos consisten de los intervalos en los cuales agrupamos dichos datos. Estos son datos son continuos y por eso representan rangos.


Por otro lado, en el eje de las y (u ordenadas) tenemos más opciones, dependiendo estas opciones es el tipo de histograma que tenemos. Los dos tipos principales de histogramas son los siguientes:

**Histograma de frecuencias absolutas**. Representa la frecuencia absoluta mediante la altura de las barras.

**Histograma de frecuencias relativas**.Representa la frecuencia relativa mediante la altura de las barras.

## **4. ¿Cómo se diferencia de un gráfico de barras?**

La diferencia fundamental entre histogramas y diagramas de barras es el tipo de datos que se representan. Los histogramas se usan en datos continuos, mientras que los diagramas de barras se usan en datos categóricos o nominales.  

Los histogramas no tienen espacio entre barras. Las barras representan la cantidad de valores que se dan en un rango específico del eje horizontal.  Los diagramas de barras tienen espacio entre barras. Las barras representan los valores de medida de cada categoría.

## **a) Definición teórica:**

La distribución de Poisson es una distribución de probabilidad discreta que modeliza la frecuencia de eventos determinados durante un intervalo de tiempo fijado a partir de la frecuencia media de aparición de dichos eventos.

**En otras palabras**, la distribución de Poisson es una distribución de probabilidad discreta que, tan solo conociendo los eventos y su frecuencia media de ocurrencia, podemos saber su probabilidad.

## **B) Condiciones para su aplicación**

Para que una distribución sea considerada como distribución de Poisson debe cumplir con tres requisitos:

--- La variable discreta “x” es el número de ocurrencias de un evento durante un intervalo determinado (de tiempo, espacio, etc.).

--- Las ocurrencias deben ser aleatorias y no contener ningún factor que favorezca unas ocurrencias en favor de otras.

--- Las ocurrencias deben estar uniformemente distribuidas dentro del intervalo que se emplee.

## **a)Definición**

La distribución binomial es una distribución de probabilidad discreta que nos dice el porcentaje en que es probable obtener un resultado entre dos posibles al realizar un número n de pruebas.

Para cada prueba, son posibles dos tipos de resultados:
A (éxito) y A* (fracaso)
La probabilidad de éxito (π) es la misma en cada prueba: probabilidad constante de éxito π (probabilidad de fracaso=1-π)
Se repite la prueba elemental un número fijo n

## **b) Propiedades**
Para que una variable aleatoria se considere que sigue una distribución binomial, tiene que cumplir las siguientes propiedades:

- en cada ensayo, experimento o prueba solo puede haber dos posibles resultados (éxito o fracaso).

- La probabilidad de éxito ha de ser constante. Por ejemplo, la probabilidad de que salga un número par al lanzar un dado es 0,5 y esta es constante dado que el dado no cambia en cada ensayo y las probabilidades de sacar par es constate.

- La probabilidad de fracaso ha de ser también constate.

- Cada experimento es independiente de los demás y no influye en las probabilidades de los que hagamos posteriormente, por lo que en cada uno la probabilidad de que se dé uno de los dos resultados será exactamente la misma.
Los sucesos son mutuamente excluyentes, es decir, no pueden ocurrir los dos al mismo tiempo. Al lanzar un dado, no puede salir par e impar a la vez, ni al lanzar una moneda puede salir cara y cruz al mismo tiempo.

- Los resultados son colectivamente exhaustivos, es decir, al menos uno de los 2 ha de ocurrir. Si no sale par, sale impar, y si no sale cara, sale cruz.

- La variable aleatoria que sigue una distribución binomial se suele representar así: X ~ (n, p), donde, n representa el número de ensayos o experimentos y p la probabilidad de éxito.
